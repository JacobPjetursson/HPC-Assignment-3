```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_data(file):
    data = pd.read_csv(file, sep="\t", header=None)
    data.columns= ["MFlops", "Time", "Memory", "Iterations", "N"]
    return data
```

```python
cpu_1 = get_data("jacobi_cpu_1.txt")
cpu_12 = get_data("jacobi_cpu_12.txt")
cpu_24 = get_data("jacobi_cpu_24.txt")
gpu_seq = get_data("jacobi_gpu_seq.txt")
gpu_naive = get_data("jacobi_gpu_naive.txt")
gpu_multi = get_data("jacobi_gpu_multi.txt")
all_data = [cpu_1, cpu_12, cpu_24, gpu_seq, gpu_naive, gpu_multi]
names = ["CPU 1", "CPU 12", "CPU 24", "GPU Seq", "GPU Naive", "GPU Multi"]
markers = ['o', '<', 's', 'P', '*', 'x']
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,5))

for i in range(len(all_data)):
    ax.plot(list(all_data[i]["N"]), list(all_data[i]["Time"]/all_data[i]["Iterations"]*1000), label=names[i], marker=markers[i], linewidth = 2.5)

ax.set_title("Execution time of one iteration", fontsize=20)
ax.set_xlabel("N", fontsize=14)
#ax.set_xlim(left=1)
ax.legend()
ax.set_ylabel("Time in milliseconds", fontsize=14)
ax.set_yscale('log')
#ax.set_xscale('log')

fig.tight_layout()

plt.savefig("Iteration_time.png", transparent=True)

plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,5))

for i in range(len(all_data)):
    ax.plot(list(all_data[i]["N"]), list(all_data[i]["MFlops"] / 1000), label=names[i], marker=markers[i], linewidth = 2.5)

ax.set_title("Gigaflops", fontsize=20)
ax.set_xlabel("N", fontsize=14)
#ax.set_xlim(left=1)
ax.legend()
ax.set_ylabel("Gigaflops", fontsize=14)
#ax.set_yscale('log')
#ax.set_xscale('log')

fig.tight_layout()

plt.savefig("Giga_flops.png", transparent=True)

plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,5))
ax.plot(list(gpu_multi["N"]), list((cpu_24["Time"]/cpu_24["Iterations"]) / (gpu_multi["Time"]/gpu_multi["Iterations"])), label="Multi CPU to multi GPU", linewidth=2.5, marker=markers[0])
ax.plot(list(gpu_naive["N"]), list((cpu_12["Time"]/cpu_12["Iterations"]) / (gpu_naive["Time"]/gpu_naive["Iterations"])), label="CPU to GPU", linewidth=2.5, marker=markers[1])
#ax.plot(list(gpu_multi["N"]), list((cpu_12["Time"]/cpu_12["Iterations"]) / (gpu_multi["Time"]/gpu_multi["Iterations"])), label="CPU to multi-GPUs", linewidth=2.5, marker=markers[2])
ax.plot(list(gpu_naive["N"]), list((cpu_24["Time"]/cpu_24["Iterations"]) / (gpu_naive["Time"]/gpu_naive["Iterations"])), label="Multiple CPU to GPU", linewidth=2.5, marker=markers[3])
#ax.plot(list(gpu_seq["N"]), list((gpu_seq["Time"]/gpu_seq["Iterations"]) / (cpu_1["Time"]/cpu_1["Iterations"])), label="GPU thread vs. CPU thread", linewidth=2.5, marker=markers[1])
#ax.plot(list(gpu_seq["N"]), list((cpu_1["Time"]/cpu_1["Iterations"]) / (gpu_seq["Time"]/gpu_seq["Iterations"]) ), label="CPU thread vs. GPU thread", linewidth=2.5, marker=markers[1])

x_1 = (cpu_24["Time"]/cpu_24["Iterations"]) / (cpu_12["Time"]/cpu_12["Iterations"])
x_2 = (gpu_multi["Time"]/gpu_multi["Iterations"]) / (gpu_naive["Time"]/gpu_naive["Iterations"])

x = list(x_2/x_1)

ax.plot(list(gpu_naive["N"]), x, label="CPUs to GPUs", linewidth=2.5, marker=markers[3])


ax.set_title("Speed-up going from CPU(s) to GPU(s)", fontsize=20)
ax.set_xlabel("N", fontsize=14)
ax.set_ylabel("Speed-up", fontsize=14)
ax.legend()
fig.tight_layout()

plt.savefig("Speedup.png", transparent=True)
plt.show()
```
